```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
countypop = "Census_Data/County Population.csv"
census = "Census_Data/ACS_17_5YR_Counties.csv"
state = "Census_Data/ACS_17_5YR_State.csv"

state_pd = pd.read_csv(state)
```

```python
#create lists for income information
education = ["Total Median Earning Less than HS", "Total Median Earning HS", "Total Median Earning Assoc", "Total Median Earning Bach", "Total Median Earning Grad"]
median_income = [24316, 29594, 33770, 47667, 60053 ]
```

```python
#plot education and median income information
plt.bar(education, median_income, color="lightcoral", align="center")
plt.title("Median Income by Education")
plt.xlabel("Education Level")
plt.ylabel("Median Income")
tick_locations = [value for value in education]
plt.xticks(tick_locations, ["Less than HS  ", "High School", "Assoc", "Bachelors", "Graduate"]);
```

```python
#percent 0f total median income by education level


colors = ["gold", "lightskyblue", "lightcoral", "pink", "cadetblue"]
labels = ["Less than HS", "High School", "Associate", "Bachelors", "Graduate"]
explode = (0, 0, 0, 0, .1)
plt.pie(median_income, explode=explode, labels=labels, colors=colors,
        autopct="%1.1f%%", shadow=True, startangle=140);

plt.title("% of Median Income by Education Level",fontsize=15)
plt.axis("equal")
t = "Note: Graph only represents ages 25+"
plt.text(-1, -1.5, t, fontsize=10, style='oblique', ha='left')

plt.show()
```
